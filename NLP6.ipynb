```python
import pandas as pd
import spacy
import nltk
from nltk.stem import PorterStemmer
```

```python
nltk.download('punkt')
```

```python
df = pd.read_csv('/content/twitter4000.csv')
display(df.head())
```

```python
nlp = spacy.load("en_core_web_sm")
stemmer = PorterStemmer()
```

```python
sample_text = df['twitts'].iloc[0]
print(sample_text)
doc = nlp(sample_text)
```

```python
tokens = [token.text for token in doc]
print(tokens)
```

```python
pos_tags = [(token.text, token.pos_) for token in doc]
print(pos_tags)
```

```python
lemmas = [(token.text, token.lemma_) for token in doc]
print(lemmas)
```

```python
stems = [(token.text, stemmer.stem(token.text)) for token in doc]
print(stems)
```

```python
entities = [(ent.text, ent.label_) for ent in doc.ents]
print(entities)
```

```python
swr = [token.text for token in doc if not token.is_stop]
print(swr)
```
